# **Feature Enginerring**

```python

df['CKD_to_ASTHMA'] = df['CKD_Perc'] / df['AST_Perc']
df['CKD_to_Diabetic'] = df['CKD_Perc'] / df['DM_Perc']
df['CKD_to_BloodPressure'] = df['CKD_Perc'] / df['HYP_Perc']
df['ASTHMA_DM_Interaction'] = df['AST_Perc'] * df['DM_Perc']
df['log_CKD_Perc'] = np.log1p(df['CKD_Perc'])
```

```python
thresh = 0.5
df.dropna(axis=1, thresh=thresh, inplace=True)
print(df)
```

```python
plt.hist(df['log_CKD_Perc'])
plt.title('Distribution of Log Transformed CKD Percentage')
plt.show()
```

ASTHMA_DM_Interaction is moderately correlated with both AST_Perc (0.77) and DM_Perc (0.82), suggesting that the combined effect of Asthma and Diabetes might be related to the individual prevalence of each condition.
Moderate Positive Correlations:

CKD_Perc is moderately correlated with AST_Perc (0.64), HYP_Perc (0.76).
AST_Perc is moderately correlated with HYP_Perc (0.73).
Weak/No Correlations:

There are weak or no correlations between DM_Perc and the other variables, except for the moderate correlation with ASTHMA_DM_Interaction (0.82).

# **Traning**

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

# ... (Your data loading and NaN removal steps) ...

# 1. Feature Selection
selected_features = [
    'All_Ages', 'N_0_3', 'N_4_11', 'N_12_18', 'N_19_29',
    'N_30_39', 'N_40_49', 'N_50_64', 'N_65_74', 'N_75_84',
    'N_85plus', 'AST_Perc', 'DM_Perc', 'HYP_Perc', 'ASTHMA_DM_Interaction','CKD_to_ASTHMA', 'CKD_to_Diabetic', 'CKD_to_BloodPressure'
]
X = df[selected_features]
y = df['CKD_Perc']  # Target variable

# 2. Identify Categorical and Numerical Features
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

# 3. Create Preprocessing Pipeline
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Model 1: Linear Regression**

```python
model_lr = LinearRegression()
pipeline_lr = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', model_lr)])
pipeline_lr.fit(X_train, y_train)
scores_lr = cross_val_score(pipeline_lr, X, y, cv=5, scoring='neg_mean_squared_error')
print("Linear Regression: Mean MSE = {:.4f} (+/- {:.4f})".format(-scores_lr.mean(), scores_lr.std()))
```

**Model 2: Random Forest (with hyperparameter tuning)**

```python
# Model 2: Random Forest (with hyperparameter tuning)
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None]
}
model_rf = GridSearchCV(RandomForestRegressor(random_state=42, n_jobs=-1), param_grid_rf, cv=5, scoring='neg_mean_squared_error')

# Correct pipeline construction
pipeline_rf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', model_rf)  # No need to double-nest the model under 'regressor'
])

pipeline_rf.fit(X_train, y_train)

# Access attributes from the GridSearchCV object
print("Random Forest: Mean MSE = {:.4f} (+/- {:.4f})".format(-model_rf.best_score_, model_rf.cv_results_['std_test_score'][model_rf.best_index_]))
```

**Model 3: Gradient Boosting (with hyperparameter tuning)**

```python
param_grid_gb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.1, 0.01]
}
model_gb = GridSearchCV(GradientBoostingRegressor(random_state=42), param_grid_gb, cv=5, scoring='neg_mean_squared_error')
pipeline_gb = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', model_gb)])
pipeline_gb.fit(X_train, y_train)
print("Gradient Boosting: Mean MSE = {:.4f} (+/- {:.4f})".format(-model_gb.best_score_, model_gb.cv_results_['std_test_score'][model_gb.best_index_]))
```

**Model 4: Support Vector Machine**

```python
model_svr = SVR(kernel='linear')  # You can try other kernels
pipeline_svr = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', model_svr)])
pipeline_svr.fit(X_train, y_train)
scores_svr = cross_val_score(pipeline_svr, X, y, cv=5, scoring='neg_mean_squared_error')
print("Support Vector Machine (Linear): Mean MSE = {:.4f} (+/- {:.4f})".format(-scores_svr.mean(), scores_svr.std()))
```

Lowest MSE: Gradient Boosting has the lowest mean MSE (0.0006), indicating that its predictions are, on average, closer to the true CKD_Perc values than the other models.
Consistent Performance: The standard deviation of the MSE for Gradient Boosting is also the lowest (0.0007), suggesting that its performance is more consistent across different folds of the cross-validation.
Other Considerations:

Random Forest: While Random Forest has a slightly higher MSE, it's still a strong contender. Its simplicity and interpretability (you can inspect feature importance) might make it a preferable choice in some situations.
Linear Regression and SVM: Both linear regression and SVM (with a linear kernel) perform significantly worse than the ensemble methods in this case. This suggests that the relationship between your features and CKD prevalence might be non-linear, and these models cannot capture that complexity.

# Testing

```python
# Evaluate Linear Regression on Test Set
y_pred_lr = pipeline_lr.predict(X_test)
mse_test_lr = mean_squared_error(y_test, y_pred_lr)
r2_test_lr = r2_score(y_test, y_pred_lr)
print(f'Linear Regression Test Set: MSE={mse_test_lr:.4f}, R-squared={r2_test_lr:.4f}')

# Linear Regression doesn't have feature_importances_
print("\nLinear Regression does not provide feature importances.")
```

```python
# Get the best model from GridSearchCV and evaluate on the test set
best_rf_model = pipeline_rf.named_steps['regressor'].best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
mse_test_rf = mean_squared_error(y_test, y_pred_rf)
r2_test_rf = r2_score(y_test, y_pred_rf)
print(f'Random Forest Test Set: MSE={mse_test_rf:.4f}, R-squared={r2_test_rf:.4f}')

# Feature Importance
feature_names = pipeline_rf.named_steps['preprocessor'].get_feature_names_out()
feature_importance_rf = pd.Series(best_rf_model.feature_importances_, index=feature_names)
print("\nRandom Forest Feature Importance:")
print(feature_importance_rf.sort_values(ascending=False))
```

```python
# Final Evaluation on Test Set
y_pred = pipeline_gb.predict(X_test)
mse_test = mean_squared_error(y_test, y_pred)
r2_test = r2_score(y_test, y_pred)
print(f'Gradient Boosting Test Set: MSE={mse_test:.4f}, R-squared={r2_test:.4f}')

# Feature Importance
best_gb_model = pipeline_gb.named_steps['regressor'].best_estimator_  # Access the best estimator
importances = best_gb_model.feature_importances_                     # Get feature importances
feature_names = pipeline_gb.named_steps['preprocessor'].get_feature_names_out()
feature_importance = pd.Series(importances, index=feature_names)
print("\nFeature Importance:")
print(feature_importance.sort_values(ascending=False))
```

```python
# Evaluate Support Vector Machine on Test Set
y_pred_svr = pipeline_svr.predict(X_test)
mse_test_svr = mean_squared_error(y_test, y_pred_svr)
r2_test_svr = r2_score(y_test, y_pred_svr)
print(f'Support Vector Machine Test Set: MSE={mse_test_svr:.4f}, R-squared={r2_test_svr:.4f}')

# SVM (Linear kernel) doesn't have feature_importances_
print("\nSupport Vector Machine does not provide feature importances.")
```
